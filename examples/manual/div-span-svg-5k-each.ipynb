```python
from IPython.display import HTML

def output_many_nodes(tag: str, n=50000):
    display(HTML(
        ''.join([
            f'<{tag}></{tag}>'
            for _ in range(n)
        ])
    ))
```

```python
output_many_nodes('div')
```

```python
output_many_nodes('span')
```

```python
output_many_nodes('svg')
```
